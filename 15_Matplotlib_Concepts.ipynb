## Creating Multiplots on Same Canvas

```python
plt.subplot(1,2,1)
# This line is creating a subplot with 1 row, 2 columns, and the current plot is set to the first one. The parameters 1, 2, 1 can be interpreted as plt.subplot(rows, columns, current_plot).

plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

## Object Oriented Method
The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Creating Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height (range 0 to 1)

# left: The starting position of the left edge of the Axes.
# bottom: The starting position of the bottom edge of the Axes.
# width: The width of the Axes.
# height: The height of the Axes.


# Plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel('Set X label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes
# So above numbers are like pixels in JavaScripts. They are just to adjust height, width of the figure

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X lable axes1')
axes1.set_ylabel('Y label axes1')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X label axes2')
axes2.set_ylabel('Y label axes2')
axes2.set_title('Axes 2 title')
```

## subplots()

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=2)
# defining number of rows and columns
```

```python
axes
# This is an array of axes to plot on. 
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig
```

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig    
plt.tight_layout()
```
