```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r') # 'r' is the color
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title Here')
plt.title('String Title')
plt.show()
```

## Creating Multiplots on Same Canvas

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(x, y, 'g*-')
```
